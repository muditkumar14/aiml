```python
import numpy as np
import pandas as pd
import requests
import json
```

```python
ur='https://restcountries.com/v3.1/all?fields=name,population,region,area'
try:
  response=requests.get(url=ur)

  print(f'HTTP Status Code:{response.status_code}')
  type(response)
  if response.status_code==200:
    countries_data=response.json()
    print(f'Total countries retrieved:{len(countries_data)}')
    print('success')
  else:
    print('error{response.status_code}')
except Exception as e:
  print(f'error{e}')
```

task 2


```python
if response.status_code==200:
  country_list=[]
  for country in countries_data:
    country_info={'common_name':country['name']['common'],
                  'population':country.get('population'),
                  'region':country.get('region'),
                  'area':country.get('area')}
    country_list.append(country_info)

  df=pd.DataFrame(country_list)
  print(f'first 10 rows {df.head(10)}')
```

Task 3

```python
pop=df.groupby('region')['population'].sum().idxmax()
print(f'Highest population :{pop}')
```

```python
max=df['area'].max()
largest=df[df['area']==max]['common_name'].values
print(f'the largest country is {largest}')
```
