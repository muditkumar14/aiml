```python
students_data = {
    'student_id': [101, 102, 103, 104, 105, 106, 107],
    'name': ['Alice', 'Bob', None, 'David', 'Emma', 'Frank', 'Grace'],
    'email': ['alice@email.com', 'bob@email.com', 'charlie@email.com', None, 'emma@email.com', 'frank@email.com', 'grace@email.com'],
    'city': ['Mumbai', 'Delhi', 'Bangalore', 'Mumbai', None, 'Chennai', 'Delhi']
}
```

```python
enrollments_data = {
    'student_id': [101, 102, 103, 105, 108, 109],
    'course_name': ['Python', 'Data Science', 'Python', 'Machine Learning', 'AI', 'Python'],
    'enrollment_date': ['2024-01-15', '2024-01-20', '2024-02-01', '2024-02-10', '2024-02-15', '2024-03-01']
}
```

```python
scores_data = {
    'student_id': [101, 102, 104, 105, 106],
    'exam_score': [85, 92, 78, 88, 95]
}
```

TASK 1


```python
import numpy as np
import pandas as pd

Students= pd.DataFrame(students_data)
Enrollments=pd.DataFrame(enrollments_data)
Scores=pd.DataFrame(scores_data)
```

```python
print("=== NULL VALUE ANALYSIS ===")
null_counts = Students.isnull().sum()
null_percentages = (null_counts / len(Students)) * 100

for column in Students.columns:
    print(f"Column: {column}, Nulls: {null_counts[column]} ({null_percentages[column]:.2f}%)")
```

```python
Students['city']=Students['city'].fillna('Unknown')
```

```python
Students=Students.dropna(subset='name')
```

```python
Students
```

TASK 2

```python
pd.merge(Students,Enrollments,on='student_id',how='inner')
```

Q.How many students appear in the result?
A.3

Q.Which students from the students table are excluded and why?
A.3 as we have done innerjoin and 103 is drop as name is NONE and 104,106 and 107 as not matched.

```python
pd.merge(Students,Enrollments,on='student_id',how='left')
```

Q.How many total rows are in the result?
A. 6

Q.Which students have null values in course_name and why?
A.3 as when we do left join all the left data is selected and doen't matter if the matched data is on right side.[104, 106, 107]

```python
pd.merge(Students,Enrollments,on='student_id',how='right')
```

Q.How many total rows are in the result?
A.6

Q.Which student_ids appear in the result but don't have student names?
A. 103,108,109

```python
pd.merge(Students,Enrollments,on='student_id',how='outer')
```

Q.How many total rows are in the result?
A.9
