```python
import requests
import json
import time
```

```python
categories = {
    'technology': ['AI', 'software', 'tech', 'code', 'computer', 'data', 'cloud', 'API', 'GPU', 'LLM'],
    'worldnews': ['war', 'government', 'country', 'president', 'election', 'climate', 'attack', 'global'],
    'sports': ['NFL', 'NBA', 'FIFA', 'sport', 'game', 'team', 'player', 'league', 'championship'],
    'science': ['research', 'study', 'space', 'physics', 'biology', 'discovery', 'NASA', 'genome'],
    'entertainment': ['movie', 'film', 'music', 'Netflix', 'game', 'book', 'show', 'award', 'streaming']
}
```

Task 1- Make the API Calls

```python
headers = {"User-Agent": "TrendPulse/1.0"}
url='https://hacker-news.firebaseio.com/v0/topstories.json'
try:
  response=requests.get(url=url,headers=headers)
  story=response.json()[:500]
  print(f"fetched {len(story)} stories")
  print(f"First 5 IDs:{story[:5]}")
except Exception as e:
  print(f"Failed to fetch{e}")
  exit()
```

Task 2- Extract the Fields

```python
category_counts = {cat: 0 for cat in categories}
collected_stories = []

for s in story:
  try:
    story_url = f'https://hacker-news.firebaseio.com/v0/item/{s}.json'
    story_response = requests.get(story_url, headers=headers)
    story_data = story_response.json()
    title = story_data.get('title', '').lower()
    category = None

    for cat, keywords in categories.items():
        if any(keyword.lower() in title for keyword in keywords):
            category = cat
            break

    if category and category_counts[category] < 25:
        category_counts[category] += 1

        story_entry = {
            "post_id": s,
            "title": story_data.get("title", ""),
            "category": category,
            "score": story_data.get("score", 0),
            "num_comments": story_data.get("descendants", 0),
            "author": story_data.get("by", "unknown"),
            "collected_at": time.strftime("%Y-%m-%d %H:%M:%S")
        }
        collected_stories.append(story_entry)

        print(f"Fetched story {s} = {category}")

        if category_counts[category] == 25:
            print(f"Category '{category}' complete. Sleeping 2 seconds\n")
            time.sleep(2)

  except Exception as e:
      print(f" Failed to fetch story {s}: {e}")
      continue

print("\n API calls complete!")
```

Task 3- Save to a JSON File

```python
import os

os.makedirs("data", exist_ok=True)
with open("data/trends_20240406.json", 'w') as f:
    json.dump(collected_stories, f, indent=2)
print(f"Collected {len(collected_stories)} stories. Saved to data/trends_20240406.json")
```
