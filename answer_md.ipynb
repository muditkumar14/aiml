```python
import requests
import os
import time # Import time for time.sleep
from typing import List

# Load API key from environment variable (never hardcode credentials)
API_URL = "https://healthstats-api.example.com/records"
API_KEY = os.getenv("HEALTHSTATS_API_KEY")  # Secure credential management

if not API_KEY:
    raise ValueError("API key not found. Please ensure 'HEALTHSTATS_API_KEY' is set in Colab secrets and retrieved by running the secrets-loading cell.")

records = []
max_pages = 10  # Respect API rate limits and usage policies

for page in range(1, max_pages + 1):
    try:
        response = requests.get(
            API_URL,
            params={"page": page, "key": API_KEY},
            timeout=30  # Add timeout to prevent hanging
        )
        response.raise_for_status()  # Check for HTTP errors
        data = response.json()

        # Only store temporary records for immediate processing
        # (not permanent storage without explicit consent/compliance review)
        records.extend(data["results"])

        # Add delay between requests to respect API rate limits
        if page < max_pages:
            time.sleep(1)

    except requests.exceptions.RequestException as e:
        print(f"Error fetching page {page}: {e}")
        break

# Process records immediately rather than permanent storage
# save_to_database(records)  # Commented out - requires ethical review first
```
