```python
import pandas as pd
import random

random.seed(42)

# Define data parameters
regions = ['North', 'South', 'East', 'West']
categories = ['Electronics', 'Clothing', 'Home & Garden', 'Sports', 'Books']
salespersons = ['Alice', 'Bob', 'Carol', 'David', 'Emma', 'Frank']

# Generate 200 sales transactions
data = {
    'transaction_id': range(1001, 1201),
    'region': [random.choice(regions) for _ in range(200)],
    'category': [random.choice(categories) for _ in range(200)],
    'salesperson': [random.choice(salespersons) for _ in range(200)],
    'sales_amount': [round(random.uniform(50, 5000), 2) for _ in range(200)],
    'customer_id': [random.randint(5000, 5100) for _ in range(200)]
}

df = pd.DataFrame(data)
print(df.head(10))
print(f"\nDataset shape: {df.shape}")

```

Task 1


```python
sale=df.groupby('region')['sales_amount'].sum().reset_index()
```

```python
df.groupby('category')['transaction_id'].count().reset_index()
```

```python
df.groupby('salesperson')['sales_amount'].mean().reset_index()
```

```python
sale.sort_values(by='sales_amount' ,ascending=False)
```

Task 2


```python
df.groupby(['region','category'])['sales_amount'].sum().reset_index()
```

```python
df.groupby('salesperson')['sales_amount'].agg(['sum','mean','count']).reset_index()
```

```python
salesperson_total=df.groupby('salesperson')['sales_amount'].agg('sum').reset_index()
```

```python
salesperson_total.sort_values(by='sales_amount',ascending=False)
```

```python
category_sale= df.groupby('category')['sales_amount'].agg('sum')
```

```python
revenue=category_sale.idxmax()
revenue
```

Task 3

```python
def sales_range(x):
    return x.max() - x.min()

```

```python
regional_sales=df.groupby('region')['sales_amount'].agg(['sum', 'mean', 'min', 'max', sales_range]).reset_index()
regional_sales
```

```python
final_summary=df.groupby('region').agg(total_number_of_transactions=('customer_id','count'),
                                       total_sales_amount=('sales_amount','sum'),
                                       average_transaction_value=('sales_amount','mean'),).reset_index()
final_summary
```

```python
df.groupby('region').agg({
    'sales_amount': ['sum', 'mean'],
    'customer_id': 'count'
}).reset_index()
```
