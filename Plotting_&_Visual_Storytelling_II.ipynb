```python
import pandas as pd
import numpy as np

np.random.seed(0)
n = 200

df = pd.DataFrame({
    'age':      np.random.randint(18, 60, n),
    'income':   np.random.randint(20000, 120000, n),
    'spend':    np.random.randint(500, 8000, n),
    'category': np.random.choice(['Electronics', 'Clothing', 'Groceries', 'Sports'], n)
})
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Task 1 — Bar Chart: Average Spend by Category

```python
avg_spend = df.groupby("category")["spend"].mean()
plt.bar(avg_spend.index, avg_spend.values)
plt.xlabel('Category')
plt.ylabel('Average Spend')
plt.title('Average Spend by Category')
plt.show()
```

Electronics has the highest average spend.

Task 2 — Scatter Plot: Income vs Spend


```python
sns.scatterplot(df, x="income", y="spend")
plt.xlabel('Income')
plt.ylabel('Spend')
plt.title('Income vs Spend')
plt.show()
```

No relation is visible


Task 3 — Interactive Scatter Plot with Plotly


```python
import plotly.express as px
fig=px.scatter(df,x=df.income,y=df.spend,color='category')
plt.xlabel('Income')
plt.ylabel('Spend')
plt.title('Income vs Spend')
fig.show()
```

This plot tells more refined data on which point what is the value both for x and y axis. this plot is heavy to load but is better in visuals.
